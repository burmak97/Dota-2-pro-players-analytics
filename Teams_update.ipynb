```python
import requests
```

```python
url = 'https://api.opendota.com/api/teams'
r = requests.get(url)
with open('JSON_DATA/teams.json', 'wb') as outfile:
    outfile.write(r.content)
```

```python
import pandas as pd
```

```python
teams = pd.read_json('JSON_DATA/teams.json')
```

```python
teams = teams[teams.last_match_time>'2022-06-01']
```

```python
teams = teams.sort_values(by=['rating'], ascending=False)
```
